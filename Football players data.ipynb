```python
import matplotlib.pyplot as plt

counts = df["preferred_foot"].value_counts()

plt.figure(figsize=(6,6))
counts.plot(
    kind="pie",
    autopct=lambda p: f'{int(round(p * counts.sum() / 100))}',
    ylabel='',  
    startangle=90
)

plt.title("Foot")
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df["position"].value_counts()

plt.figure(figsize=(6,6))
counts.plot(
    kind="pie",
    autopct=lambda p: f'{int(round(p * counts.sum() / 100))}',
    ylabel='',  
    startangle=90
)

plt.title("Game position")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) 
df["team"].value_counts().plot(kind="bar")

plt.xlabel("Count")
plt.ylabel("Team")
plt.title("Team Counts")
plt.show()
```
